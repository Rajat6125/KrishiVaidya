```python
import os
import random
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
dataset_path = Path("data/raw/PlantVillage")
print(dataset_path.exists())
```

```python
classes = sorted(os.listdir(dataset_path))

print(f"Number of classes : {len(classes)}")
```

```python
image_count = {}

for i in classes:
    class_path = dataset_path / i
    image_count[i] = len(os.listdir(class_path))
```

```python
df = pd.DataFrame(image_count.items(),columns=["Class", "Images"])
df
```

```python
plt.bar(df["Class"], df["Images"])
plt.xticks(rotation=90)
plt.title("Number of Images per Class")
plt.xlabel("Class")
plt.ylabel("Images")

plt.show()
```
